# QuatPlot demo
Interactive 3D plots with quaternion rotation (via ganja.js, no gimbal lock), rendered in an iframe.

Requirements: `quatplot.py` in the same directory (or on `sys.path`), numpy, and a Jupyter environment. The ganja.js CDN is used when reachable; otherwise a built-in quaternion fallback kicks in automatically.

```python
import numpy as np
import quatplot as qp
```

## Surface plot
MATLAB-style *peaks* function on a 41×41 grid. Drag to rotate, wheel to zoom, shift-drag to pan. Try **Top view** with isometric projection for a 2D-heatmap look.

```python
x = np.linspace(-3, 3, 41)
y = np.linspace(-3, 3, 41)
X, Y = np.meshgrid(x, y)
Z = (3*(1 - X)**2 * np.exp(-X**2 - (Y + 1)**2)
     - 10*(X/5 - X**3 - Y**5) * np.exp(-X**2 - Y**2)
     - 1/3 * np.exp(-(X + 1)**2 - Y**2))

qp.plot(qp.Surface(x, y, Z), height=520)
```

## Scatter plot
Three gaussian clusters plus a helix, colored by a value array `v`. Hover a marker for its coordinates; toggle the **Color bar** to see the value scale.

```python
rng = np.random.default_rng(42)
centers = np.array([[-1.8, -1.8, -4], [2, 1.5, 2], [-1, 2.2, 6]])
cl = np.concatenate([c + rng.normal(scale=[0.5, 0.5, 1.2], size=(35, 3)) for c in centers])
cval = np.repeat([0.0, 1/3, 2/3], 35)

t = np.linspace(0, 4*np.pi, 45)
hx, hy, hz = 2.6*np.cos(t), 2.6*np.sin(t), np.linspace(-6, 6, 45)
hv = 0.6 + 0.4*np.linspace(0, 1, 45)

px = np.concatenate([cl[:, 0], hx]); py = np.concatenate([cl[:, 1], hy])
pz = np.concatenate([cl[:, 2], hz]); pv = np.concatenate([cval, hv])

qp.plot(qp.Scatter3(px, py, pz, v=pv), colorbar=True, cmap='Plasma', height=520)
```

## Combined: surface + scatter
Multiple traces overlay in one plot; axes autoscale to the union of all data.

```python
sx = rng.uniform(-3, 3, 60)
sy = rng.uniform(-3, 3, 60)
sz = (3*(1 - sx)**2 * np.exp(-sx**2 - (sy + 1)**2)
      - 10*(sx/5 - sx**3 - sy**5) * np.exp(-sx**2 - sy**2)
      - 1/3 * np.exp(-(sx + 1)**2 - sy**2)) + 1.5   # samples floating above the surface

qp.plot(qp.Surface(x, y, Z), qp.Scatter3(sx, sy, sz), theme='light', height=520)
```

## Standalone HTML export
`qp.save` writes a self-contained file — same template, and the Screenshot button downloads directly there.

```python
qp.save('peaks.html', qp.Surface(x, y, Z), cmap='Coolwarm')
```
